# pandas.DataFrame.loc
#### property DataFrame.loc
Access a group of rows and columns by label(s) or a boolean array.

> df.loc[row_sel, col_sel]

이름을 이용하여 원하는 row 에서 원하는 col 선택

```python
df.loc['No.1'] # row_sel
```

```python
df.loc['No.5','Python'] # row_sel, col_sel
```

```python
df.loc[['No.1','No.2'], 'School'] # row_sel, col_sel
```

```python
df.loc[['No.1', 'No.6'], ['School','Gender']] # row_sel, col_sel
```

```python
df.loc['No.2':'No4', 'Python':'HTML'] # row_sel, col_sel
```

***
# pandas.DataFrame.iloc
#### property DataFrame.iloc
Purely integer-location based indexing for selection by position.

> df.iloc[row_sel, col_sel]

위치를 이용하여 원하는 row 에서 원하는 col 선택

```python
df.iloc[0] # row_sel
```

```python
df.iloc[4] # row_sel
```

```python
df.iloc[1:3] # row_sel
```

```python
df.iloc[2, 3] # row_sel, col_sel
```

```python
df.iloc[[2, 3], 4] # row_sel, col_sel
```

```python
df.iloc[[2, 3], [4, 5]] # row_sel, col_sel
```

```python
df.iloc[:3, 1:5] # row_sel, col_sel
```

## 조건에 해당하는 값 선택

```python
df['Height'] > 180
```

```python
filt = df['Height'] > 180 
df[filt] # No.2, No.3, No.7 (True) 값 만을 반환
```

```python
df[~filt] # 2, 3, 7 이외의 값 반환
```

```python
df.loc[filt, 'Python'] # row_sel, col_sel
```

```python
df.loc[filt, ['School','Gender','Python']] # row_sel, col_sel
```

```python
filt1 = df['Height'] > 180
filt2 = df['School'] == 'MIT'
```

### 조건 : &

```python
df.loc[filt1 & filt2] # 특정 조건(filtering)을 만족하는 데이터 찾기
```

### 조건 : |

```python
df.loc[filt1 | filt2]
```

```python
filt = df['School'].str.startswith('M') # M으로 시작하는 학교를 다니는 사람에 대한 데이터
df[filt]
```

```python
df[~filt] # M으로 시작하는 학교를 다니는 사람 제외한 데이터
```

```python
sch = ['MIT','Harvard']
filt = df['School'].isin(sch) # 학교가 MIT 또는 Harvard인 사람에 대한 데이터
df[filt]
```

```python
df['Name'].str.upper() # 모두 대문자로 표현
```
